```python
import pandas as pd
from scipy.stats import chi2_contingency

# Function to perform the Chi-Square Test
def perform_chi_square_test(df, section_col, dv_col):
    # Creating a contingency table
    contingency_table = pd.crosstab(df[section_col], df[dv_col])

    # Performing the Chi-Square Test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Interpreting the result
    significant = p < 0.05  # 5% significance level
    return chi2, p, significant

# Calling the function
chi2_stat, p_value, is_significant = perform_chi_square_test(df, "SECTION", "DV")

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Significant? {'Yes' if is_significant else 'No'}")
```

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

# Apply the function to the 'Score' column
df['Grade'] = df['percentage'].apply(assign_grade)

print(df)
```
