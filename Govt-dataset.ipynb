```python
df['town_municipalty_name'].fillna("Unknown", inplace=True)
df['ward_name'].fillna("Unknown", inplace=True)
df['basin_name'].fillna("Unknown", inplace=True)
df['sub_basin_name'].fillna("Unknown", inplace=True)
df['no_town_cities_benefited'].fillna("Unknown", inplace=True)
df['reason_water_body_in_use_name2'].fillna("Unknown", inplace=True)
df['reason_water_body_in_use_name3'].fillna("Unknown", inplace=True)
df['no_villages_benefited'].fillna("Unknown", inplace=True)
df['ref_selection_id_wua_exists_name'].fillna("Unknown", inplace=True)
df['Block/Tehsil Name'] = df['Block/Tehsil Name'].fillna("Unknown")
df['Village Name'] = df['Village Name'].fillna("Unknown")
```

Replacing Categorical values with unknown 

```python
threshold = 90
df = df.loc[:, df.isnull().mean() * 100 < threshold]  
```

```python
df['longitude_dec'].fillna(df['longitude_dec'].median(), inplace=True)
df['latitude_dec'].fillna(df['latitude_dec'].median(), inplace=True)
df['ipc_water_body'].fillna(df['ipc_water_body'].median(), inplace=True)
df['cca_water_body'].fillna(df['cca_water_body'].median(), inplace=True)
df['water_spread_area_of_water_body'] = df['water_spread_area_of_water_body'].fillna(df['water_spread_area_of_water_body'].median())
```

Replacing Numerical values with median

```python
df
```

```python
df['no_people_benefited_by_water_body'].fillna(df['no_people_benefited_by_water_body'].mode()[0], inplace=True)
df['storage_capacity_water_body_original'].fillna(df['storage_capacity_water_body_original'].mode()[0], inplace=True)
df['storage_capacity_water_body_present'].fillna(df['storage_capacity_water_body_present'].mode()[0], inplace=True)
df['filled_up_storage_name'].fillna(df['filled_up_storage_name'].mode()[0], inplace=True)
df['filled_up_storage_space_name'].fillna(df['filled_up_storage_space_name'].mode()[0], inplace=True)
df['max_depth_water_body_fully_filled'].fillna(df['max_depth_water_body_fully_filled'].mode()[0], inplace=True)
df['construcion_year'].fillna(df['construcion_year'].mode()[0], inplace=True)
df['construction_cost'].fillna(df['construction_cost'].mode()[0], inplace=True)
df['water_body_name'].fillna(df['water_body_name'].mode()[0], inplace=True)
```

```python
import scipy.stats as stats
```

```python
sample_data = df['no_people_benefited_by_water_body'].dropna()  # Remove NaN values
t_stat, p_value = stats.ttest_1samp(sample_data, 500)
print(f"One-Sample T-Test Results: t-statistic = {t_stat}, p-value = {p_value}")
```

```python
group_rural = df[df['rural_or_urban'] == 'Rural']['no_people_benefited_by_water_body'].dropna()
group_urban = df[df['rural_or_urban'] == 'Urban']['no_people_benefited_by_water_body'].dropna()
t_stat, p_value = stats.ttest_ind(group_rural, group_urban, equal_var=False)  # Set equal_var=False if variances are different
print(f"Independent Two-Sample T-Test Results: t-statistic = {t_stat}, p-value = {p_value}")
```

```python
t_stat, p_value = stats.ttest_ind(group_rural, group_urban, equal_var=False)  
print(f"Welch’s T-Test Results: t-statistic = {t_stat}, p-value = {p_value}")
```

```python
t_stat, p_value = stats.ttest_ind(group_rural, group_urban, equal_var=False)
one_tailed_p = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)
print(f"One-Tailed T-Test p-value: {one_tailed_p}")
```
