# Rounding percentage values into two decimals

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

# Apply the function to the 'Score' column
df['Grade'] = df['percentage'].apply(assign_grade)

print(df)
```
